```python
#@title Checking for NA/NaN Values
print(df.isnull().all())
print(df.notna().all())
```

```python
#@title Checking for duplicate values
boolean_series = df.duplicated(keep='first')
print('Boolean series:')
print(boolean_series)
print('\n')
print('DataFrame after keeping only the first instance of the duplicate rows:')

df[~boolean_series]
#in conclusion there appears to be no duplicate values to worry about
```

```python
#second method
df.drop_duplicates(subset=None, keep='first', inplace=False, ignore_index=False)
```

```python
#@title Checking for outliers
#age
import plotly.express as px
df.describe()[['Dexa_Freq_During_Rx','Count_Of_Risks']]
#Boxplot Vizualization
Rx = px.box(df, y= "Dexa_Freq_During_Rx")
Rx.show()

Risks = px.box(df, y= "Count_Of_Risks")
Risks.show()

#Multivariate outliers using scatterplot
fig = px.scatter(x=df['Count_Of_Risks'], y=df['Dexa_Freq_During_Rx'])
fig.show()

#Both 'Dexa_Freq_During_Rx' & 'Count_Of_Risks' have outliers with 'Dexa_Freq_During_Rx' having sifnificantly more. This implies both are skewed
#It should be noted quantitative variables (like 'Age_Bucket') is not analyzed here as it is not in an appropriate index
```

```python
#Finding the outliers
def find_outliers(df):
   q1=df.quantile(0.25)
   q3=df.quantile(0.75)
   IQR=q3-q1
   outliers = df[((df<(q1-1.5*IQR)) | (df>(q3+1.5*IQR)))]
   return outliers
   
print("Finding outliers for 'Dexa_Freq_During_Rx'")
outliers = find_outliers(df['Dexa_Freq_During_Rx'])
print("number of outliers: " + str(len(outliers)))
print("max outlier value: " + str(outliers.max()))
print("min outlier value: " + str(outliers.min()))
outliers
```

```python
print("Finding outliers for 'Count_Of_Risks'")
outliers = find_outliers(df['Count_Of_Risks'])
print("number of outliers: " + str(len(outliers)))
print("max outlier value: " + str(outliers.max()))
print("min outlier value: " + str(outliers.min()))
outliers
```

```python
#@title Skewness (related to outliers above) 
#Vizualiations using histograms
Rx_Skew = px.histogram(df, x='Dexa_Freq_During_Rx')
Rx_Skew.show()

Risks_Skew = px.histogram(df, x='Count_Of_Risks')
Risks_Skew.show()

df.skew(axis = 0, skipna = True)

#Both 'Dexa_Freq_During_Rx' & 'Count_Of_Risks' are skewed right with 'Dexa_Freq_During_Rx'being more severely skewed.
```
